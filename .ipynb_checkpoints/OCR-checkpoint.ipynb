```python
import pytesseract
from PIL import Image
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
_344 = cv2.imread('../input/344.jpg')
plt.imshow(cv2.cvtColor(_344,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
paper = _344.copy()
x = 200
condition = np.bitwise_or(paper[:,:,0] < x, paper[:,:,1] < x, paper[:,:,2] < x)
paper[condition] = 0
paper[np.invert(condition)] = 255
plt.imshow(cv2.cvtColor(paper,cv2.COLOR_BGR2RGB))
plt.show()
gray = cv2.cvtColor(paper, cv2.COLOR_BGR2GRAY)
```

```python
x, y = cv2.findNonZero(gray)[0][0]
#print(pytesseract.image_to_string(_344))
test = paper[y:, x:]
plt.imshow(test)
plt.show()
```

```python
print(pytesseract.image_to_string(test))
```
